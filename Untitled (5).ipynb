```python
#[1]:
#In Python, we can define custom exceptions by creating a new class that is derived from the built-in Exception class.
Example:
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass

# you need to guess this number
number = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < number:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```

```python
#[3]:
The ArithmeticError class in Python defines errors that can occur during arithmetic operations.
Two common errors in this class are:
    
ZeroDivisionError: This error is raised when attempting to divide a number by zero. 
For example:
    try:
        a=10/0
    except ZeroDivisionError as e:
        print(e)
        
OverflowError:
This error is raised when the result of an arithmetic operation is too large to be represented.
For example:
    x = 2**1000
# OverflowError: int too large to convert to float
```

```python
#[4]:
-The LookupError class is a base class for exception classes that indicate a lookup operation has failed, 
-such as when trying to access a non-existent key in a dictionary, or an out-of-bounds index in a sequence. 
-LookupError is a subclass of the built-in Exception class in Python.
--KeyError and IndexError are two subclasses of LookupError that are commonly used to indicate lookup failures for dictionaries and sequences,
respectively.
#Example of key error:

try:
    d={"key":"kissu","key1":"mira"}
    print(d["key2"])
except KeyError as e:
    print(e)
    
#Example of IndexError:

try:
    d=[1,4,7,2,8,3]
    print(d[9])
except IndexError as e:
    print(e)
```
