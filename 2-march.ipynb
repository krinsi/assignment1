```python
#[q-1]:
-Matplotlib is a data visualization library in Python that provides tools for creating a wide variety of static, animated, and interactive
 plots.
-It is a popular and widely-used library because it is flexible, customizable, and can produce high-quality visualizations.
```

```python
#name of the plot:
Line plots: 
    Line plots are used to represent the relationship between two variables on a continuous axis. 
    They can be used to show trends over time or to compare different groups.

Scatter plots:
    Scatter plots are used to represent the relationship between two variables on a Cartesian coordinate system.
    They can be used to identify patterns or clusters in the data.

Bar plots:
    Bar plots are used to represent categorical data using rectangular bars.
    They can be used to compare the values of different categories or to show changes over time.

Histograms: 
    Histograms are used to represent the distribution of a continuous variable. 
    They can be used to identify the shape of the distribution, including its central tendency and spread.

Box plots:
    Box plots, also known as box and whisker plots, are used to represent the distribution of a dataset using five statistics: the minimum 
    value, the first quartile, the median, the third quartile, and the maximum value.
```

```python
#[q-2]:
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
plt.scatter(x,y)
```

```python
#[q-3]:
The subplot() function:
    The subplot() functionin Matplotlib is used to create multiple plots within a single figure.
    It allows you to display multiple plots in a grid-like structure, where each plot can be customized individually.
```

```python
fig, axs = plt.subplots(2, 2)

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
axs[0, 0].plot(x1, y1)
plt.title('Line 1')

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Set the figure title
fig.suptitle('Four Line Plots')

# Show the plot
plt.show()
```

```python
#[q-4]:
bar plot function:
-A bar plot is a type of chart that represents data using rectangular bars.
-Each bar represents a category and its height represents the value of a variable for that category.
-Bar plots are commonly used to compare the values of different categories or to track changes in a variable over time.
-Bar plots are useful because they allow us to quickly and easily compare the values of different categories.
-They are also effective at showing the magnitude and direction of changes in a variable over time, especially when the bars are stacked or
 grouped.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
```

```python
#[q-5]:
-A box plot, also known as a box and whisker plot, is a graphical representation of a dataset that shows the distribution of the data
 using five statistics: the minimum value, the first quartile (25th percentile), the median (50th percentile), the third quartile 
 (75th percentile), and the maximum value. It also displays any outliers that may exist in the data.
-Box plots are commonly used to visually summarize the distribution of a dataset and to compare the distribution of different datasets.
-They are particularly useful when working with large datasets or datasets that contain many outliers.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot Example')
plt.show()
```
