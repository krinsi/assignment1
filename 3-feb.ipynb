```python
#[q-1]:
Scatter Plot:
    A scatter plot is a two-dimensional plot that uses dots to represent values for two different variables. 
    Seaborns scatterplot function allows us to create scatter plots with more features and options than Matplotlibs scatter function. 
    We can use scatter plots to visualize the relationship between two continuous variables.
    
Line Plot: 
    A line plot is a type of plot that displays information as a series of data points connected by straight line segments. 
    Seaborns lineplot function allows us to create line plots with more features and options than Matplotlibs plot function.
    
Histogram:
    A histogram is a graphical representation of the distribution of a dataset. 
    Seaborn histplot function allows us to create histograms with more features and options than Matplotlibs hist function.
    
Box Plot: 
    A box plot is a graphical representation of the distribution of a dataset based on five number summary (minimum, first quartile, median,
    third quartile, and maximum). 
    Seaborns boxplot function allows us to create box plots with more features and options than Matplotlibs boxplot function.
    
Heatmap:
    A heatmap is a graphical representation of data in which values are represented by colors.
    Seaborns heatmap function allows us to create heatmaps with more features and options than Matplotlibs imshow function. 
```

```python
#[q-2]:
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```
