```python
#[q-1]:
Both t-tests and z-tests are statistical hypothesis tests that help us to make inferences about the population based on a sample of data.
-However, they differ in their underlying assumptions and the situations in which they are appropriate.

-A t-test is used when the sample size is small (typically less than 30), and the population standard deviation is unknown.
-It is designed to test hypotheses about the mean of a normally distributed population when the sample size is too small to use the standard normal
 distribution.
-A t-test uses the t-distribution instead of the standard normal distribution, which takes into account the extra uncertainty that arises from the
 small sample size.

#EXAMPLE:
-Suppose you want to test the hypothesis that the mean height of male college students is 70 inches.
-You take a random sample of 25 male college students and measure their heights. 
-The sample mean height is 68 inches, and the sample standard deviation is 2.5 inches.

-In this case, you would use a t-test because the sample size is small (less than 30), and the population standard deviation is unknown. 
-You would calculate the t-statistic and compare it to the critical t-value from the t-distribution to determine whether to reject or fail to reject
the null hypothesis.
```

```python
#[q-2]:
-In hypothesis testing, a one-tailed test is a statistical test where the alternative hypothesis is directional, meaning it predicts the direction
 of the effect or relationship between variables.
-This means that the test is designed to determine whether a sample mean or proportion is either significantly greater than or significantly less
 than a hypothesized population mean or proportion.
-On the other hand, a two-tailed test is a statistical test where the alternative hypothesis is non-directional, meaning it simply predicts that
 there is a difference between the sample and the hypothesized population mean or proportion, without predicting the direction of the difference.
```

```python
#[q-3]:
-Type 1 error occurs when we reject a null hypothesis that is actually true. 
-It is also known as a false positive.
-This error occurs when we conclude that there is a significant effect or relationship between variables when in reality there is none.
-The probability of making a Type 1 error is denoted by alpha (α) and is usually set at 0.05 or 0.01.

Example scenario for Type 1 error:
    Suppose a new drug is being tested to treat a disease. 
    The null hypothesis is that the drug has no effect on the disease. 
    A Type 1 error would occur if the study concludes that the drug is effective (rejects the null hypothesis) when it is actually not effective.
```

```python
#[q-4]:
Bayess theorem is a fundamental concept in probability theory that describes how to update our beliefs about an event or hypothesis based on new evidence or data. It is named after the English statistician and philosopher, Thomas Bayes.

Bayess theorem states that the probability of a hypothesis (H) given some observed evidence (E) is equal to the product of the prior probability of the hypothesis (P(H)) and the probability of the evidence given the hypothesis (P(E|H)), divided by the probability of the evidence (P(E)):

P(H|E) = (P(E|H) * P(H)) / P(E)

Suppose a medical test is designed to detect a certain disease. The test is 95% accurate, meaning that the probability of a false positive (i.e., the test result is positive when the person does not have the disease) is 5%, and the probability of a false negative (i.e., the test result is negative when the person does have the disease) is also 5%. The prevalence of the disease in the population is 1%.

Given this information, lets use Bayess theorem to calculate the probability that a person has the disease given that the test result is positive (i.e., P(Disease|Positive Test)).

Step 1: Prior probability (P(Disease))
The prior probability of the person having the disease is 1%.

P(Disease) = 0.01

Step 2: Probability of the evidence given the hypothesis (P(Positive Test|Disease))
The probability of a positive test given that the person has the disease is 95%.

P(Positive Test|Disease) = 0.95

Step 3: Probability of the evidence (P(Positive Test))
To calculate the probability of a positive test, we need to consider both true positives (people who have the disease and test positive) and false positives (people who do not have the disease but test positive).

P(Positive Test) = P(Positive Test|Disease) * P(Disease) + P(Positive Test|No Disease) * P(No Disease)
= 0.95 * 0.01 + 0.05 * 0.99
= 0.058

Step 4: Probability of the hypothesis given the evidence (P(Disease|Positive Test))
Using Bayess theorem, we can now calculate the probability of the person having the disease given that the test result is positive:

P(Disease|Positive Test) = (P(Positive Test|Disease) * P(Disease)) / P(Positive Test)
= 0.95 * 0.01 / 0.058
= 0.163 or 16.3%

Therefore, the probability that the person has the disease given a positive test result is 16.3%. This example shows how Bayess theorem can be used to
update our beliefs about the probability of a hypothesis given new evidence or data.





```

```python
#[q-5]:
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a specified level of confidence.
It is an important statistical tool used in inferential statistics to estimate population parameters based on sample statistics.

-The confidence interval provides a measure of the uncertainty associated with an estimate and is typically expressed as a range of values with an
 associated level of confidence, usually 90%, 95%, or 99%.
-For example, a 95% confidence interval for a population mean would provide a range of values that is expected to contain the true population mean 
 with a probability of 0.95.
    
Suppose we want to estimate the average height of all students in a school. We randomly sample 100 students from the school and measure their heights. We find that the sample mean height is 170 cm, and the sample standard deviation is 5 cm.

Step 1: Determine the sample statistic
The sample statistic in this case is the sample mean height, which is 170 cm.

Step 2: Determine the standard error
The standard error is a measure of the variability of the sample mean and is calculated as the sample standard deviation divided by the square root
of the sample size:

Standard Error = Standard Deviation / Square Root of Sample Size
= 5 / √100
= 0.5
```

```python
#[q-6]:
Problem:
A doctor believes that a patient has a 40% chance of having a certain disease based on the patients symptoms.
The doctor orders a diagnostic test, which has a 90% probability of producing a positive result if the patient has the disease and an 80% probability of producing a negative result if the patient does not have the disease. If the test comes back positive, what is the probability that the patient has the disease?

Solution:
Lets use Bayes Theorem to calculate the probability of the patient having the disease given a positive test result:

Step 1: Prior probability (P(Disease))
The prior probability of the patient having the disease based on the doctors assessment is 0.4 or 40%.

P(Disease) = 0.4

Step 2: Probability of the evidence given the hypothesis (P(Positive Test|Disease))
The probability of a positive test given that the patient has the disease is 0.9 or 90%.

P(Positive Test|Disease) = 0.9

P(Positive Test) = P(Positive Test|Disease) * P(Disease) + P(Positive Test|No Disease) * P(No Disease)
= 0.9 * 0.4 + 0.2 * 0.6
= 0.5

Step 4: Probability of the hypothesis given the evidence (P(Disease|Positive Test))
Using Bayes Theorem, we can now calculate the probability of the patient having the disease given a positive test result:

P(Disease|Positive Test) = (P(Positive Test|Disease) * P(Disease)) / P(Positive Test)
= 0.9 * 0.4 / 0.5
= 0.72 or 72%
```

```python
#[q-7]:
To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, we can use the following formula:

Confidence Interval = Sample Mean ± (Critical Value x Standard Error)

where the critical value is based on the desired level of confidence and the sample size, and the standard error is calculated as the sample standard deviation divided by the square root of the sample size.

For a 95% confidence interval with a sample size of 30 or more, the critical value is 1.96. Since the sample size is not given in the problem, we will assume that it is large enough for this critical value to be used.

Step 1: Calculate the standard error
Standard Error = Standard Deviation / Square Root of Sample Size
= 5 / √n

Step 2: Find the critical value
For a 95% confidence interval, the critical value is 1.96.

Step 3: Calculate the confidence interval
Confidence Interval = Sample Mean ± (Critical Value x Standard Error)
= 50 ± (1.96 x (5 / √n))

Interpretation:
The 95% confidence interval for the population mean is 50 ± (1.96 x (5 / √n)). This means that if we take many samples of the same size from this population, and calculate the 95% confidence interval for each sample mean, then we would expect 95% of these intervals to contain the true population mean.

Since we do not know the sample size, we cannot give a specific range for the confidence interval. However, as the sample size increases, the range of the confidence interval will become smaller, indicating greater precision in our estimate of the population mean.
```

```python
#[q-8]:
The formula for the margin of error is: Margin of Error = z* x (Standard Error)

Sample size affects the margin of error in two ways:

Inversely Proportional: The margin of error is inversely proportional to the square root of the sample size. This means that as the sample size increases, the margin of error decreases. In other words, a larger sample size generally leads to a more precise estimate of the population parameter.

Directly Proportional: The margin of error is directly proportional to the standard error of the sample. This means that if the standard error of the sample is larger, the margin of error will be larger, and vice versa.

Example scenario:
Suppose we want to estimate the average age of all employees in a company with 95% confidence. We take a random sample of 50 employees and calculate their ages. Suppose the mean age of the sample is 35 years and the standard deviation is 6 years.

The margin of error for a 95% confidence interval can be calculated as:
Margin of Error = z* x (Standard Error)
where z* = 1.96 (for a 95% confidence interval)
and Standard Error = standard deviation / sqrt(sample size)

Thus, Margin of Error = 1.96 x (6 / sqrt(50)) = 1.68

This means that we can be 95% confident that the true average age of all employees in the company lies between 33.32 (35 - 1.68) and 36.68 (35 + 1.68) years.
```

```python
#[q-9]:
The z-score for a data point is a measure of how many standard deviations the data point is away from the population mean. It is calculated as:

z = (x - μ) / σ

where x is the data point, μ is the population mean, and σ is the population standard deviation.

In this case, the data point is 75, the population mean is 70, and the population standard deviation is 5. Plugging in these values, we get:

z = (75 - 70) / 5 = 1

This means that the data point of 75 is one standard deviation above the population mean.

Z-scores are used to standardize data, making it easier to compare different data points on a common scale. In this case, a z-score of 1 indicates that the data point is higher than the population mean by one standard deviation. This information can be useful in understanding how extreme or unusual the data point is in relation to the rest of the population.
```

```python
#[q-10]:
To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to set up the null and alternative hypotheses:

Null hypothesis: The weight loss drug is not significantly effective, and the population mean weight loss is equal to or less than 0 pounds. (H0: μ ≤ 0)
Alternative hypothesis: The weight loss drug is significantly effective, and the population mean weight loss is greater than 0 pounds. (Ha: μ > 0)

We will use a one-tailed t-test since we are testing whether the population mean weight loss is greater than 0. The test statistic is calculated as:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the hypothesized population mean under the null hypothesis, s is the sample standard deviation, and n is the sample size.

Plugging in the given values, we get:

t = (6 - 0) / (2.5 / √50) = 12

Using a t-table with 49 degrees of freedom (df = n - 1), we find that the critical t-value at a 95% confidence level is 1.677.

Since our calculated t-value of 12 is greater than the critical t-value of 1.677, we reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

Therefore, we can say that there is evidence to suggest that the population mean weight loss with the new drug is greater than 0 pounds.
```

```python
#[q-12]:
To conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01, we need to set up the null and alternative hypotheses:

Null hypothesis: The two teaching methods do not have a significant difference in student performance. (H0: μA - μB = 0)
Alternative hypothesis: The two teaching methods have a significant difference in student performance. (Ha: μA - μB ≠ 0)

We will use a two-tailed t-test since we are testing for a difference in means. The test statistic is calculated as:

t = (x̄A - x̄B) / (sqrt((sA^2/nA) + (sB^2/nB)))

where x̄A and x̄B are the sample means, sA and sB are the sample standard deviations, nA and nB are the sample sizes.

Plugging in the given values, we get:

t = (85 - 82) / (sqrt((6^2/1) + (5^2/1))) = 4.26

Using a t-table with 2 degrees of freedom less than the sample size (df = nA + nB - 2), we find that the critical t-value at a 0.01 significance level is ±2.576.

Since our calculated t-value of 4.26 is greater than the critical t-value of 2.576, we reject the null hypothesis and conclude that there is a significant difference in student performance between the two teaching methods at a 0.01 significance level.

Therefore, we can say that there is evidence to suggest that one teaching method is more effective than the other in improving student performance.
```
